```python
import pandas as pd
import numpy as np

caminho = r'C:\Users\leandro.rocha.A1\PycharmProjects\Linkedin_python_ds\dados\mtcars.csv'

carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
carros.head()
```

```python
carros_cat = carros[['cyl','am','gear','carb']]
```

```python
marchas_grupos = carros_cat.groupby('gear')
marchas_grupos.describe().T #esse . T traspoe colunas em linhas!
```
