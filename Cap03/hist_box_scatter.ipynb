```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns

%matplotlib inline
rcParams['figure.figsize']=5,4
sns.set_style('whitegrid')

import warnings
warnings.filterwarnings('ignore')
```

```python
caminho = (r'C:\Users\leandro.rocha.A1\PycharmProjects\Linkedin_python_ds\dados\mtcars.csv')
carros = pd.read_csv(caminho)
carros.columns = ['nome','mpg','cyl','disp','hp','draft','wt','qsec','vs','am','gear','carb']
mpg = carros['mpg']

plt.hist(mpg)
plt.show()
```

```python
sns.distplot(mpg)
```

```python
carros.plot(kind = 'scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
sns.regplot(x='hp', y='mpg', data=carros, scatter=True)
```

```python
#matrizes de scatter
sns.pairplot(carros)
```

```python
colunas = ['mpg','disp','hp','wt']
novo_df = pd.DataFrame((carros[colunas].values), columns=colunas)
carros_target = carros['am'].values

novo_df['grupo'] = pd.Series(carros_target, dtype='category')
```

```python
sns.pairplot(novo_df, hue='grupo',palette='hls')
```

```python
carros.boxplot(column='mpg', by='am')
carros.boxplot(column='wt', by='am')
```

```python
sns.boxenplot(x='am', y='mpg', data=carros, palette= 'hls')
```
