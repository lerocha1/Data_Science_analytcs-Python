```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams

import warnings
warnings.filterwarnings('ignore')
```

```python
%matplotlib inline
rcParams['figure.figsize']=8,4
sns.set_style('whitegrid')

caminho = (r'C:\Users\leand\PycharmProjects\Linkedin_python_ds\dados\mtcars.csv')
carros = pd.read_csv(caminho)
carros.columns = ['nome','mpg','cyl','disp','hp','draft','wt','qsec','vs','am','gear','carb']
mpg = carros['mpg']
```

```python
x = range(1,10)
y = [1,2,3,4,5,4,3,2,1]
plt.bar(x,y)
plt.xlabel('nosso titulo em x')
plt.ylabel('nosso titulo em y')
```

```python
import matplotlib.patches as mpatches

plt.plot(mpg)
plt.xlabel('nomes dos carros')
plt.ylabel('milhas/galo')
plt.title('Consumo de Gasolina')
plt.xticks(mpg.index, carros.nome, rotation=45)

legenda = mpatches.Patch(label='mpg')
plt.legend(handles=[legenda])
```
